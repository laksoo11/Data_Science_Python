```python
# Ensure the 'Date' column is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Sort the data by Date to ensure proper time order
data_cleaned = data_cleaned.sort_values(by='Date')

# Group the data by 'Date' and calculate the sum of sales per day
sales_by_date = data_cleaned.groupby('Date')['Sales'].sum()
sales_by_date
```

```python
# Create a line plot of Sales over time
plt.figure(figsize=(10, 5))
plt.plot(sales_by_date.index, sales_by_date.values, color='g', marker='*')

# Add titles and labels
plt.title("Sales Trend Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Createing a figure with two subplots for Price and Sales
plt.figure(figsize=(10, 6))

# Boxplot for Price
plt.subplot(1, 2, 1)
sns.boxplot(data=data_cleaned, x='Price', color='lightblue')
plt.title('Price Distribution')

# Boxplot for Sales
plt.subplot(1, 2, 2)
sns.boxplot(data=data_cleaned, x='Sales', color='lightgreen')
plt.title('Sales Distribution')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
bins = [0, 100, 200, 300, 400, 500, 1000]  # Example price ranges
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501+']  # Labels for the bins

# Create a new column for price range
data_cleaned['Price_Range'] = pd.cut(data_cleaned['Price'], bins=bins, labels=labels)

# Group by Price_Range and calculate the average sales for each range
sales_by_price_range = data_cleaned.groupby('Price_Range')['Sales'].mean()

# Display the results
print(sales_by_price_range)
```

```python
# Creating the bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=sales_by_price_range.index, y=sales_by_price_range.values, color='lightgreen')

# Add titles and labels
plt.title("Average Sales by Price Range", fontsize=15)
plt.xlabel("Price Range", fontsize=11)
plt.ylabel("Average Sales", fontsize=11)

# Display values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

# To Show the plot
plt.show()
```

```python

data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Extract Month and Day of the Week from Date
data_cleaned['Month'] = data_cleaned['Date'].dt.month
data_cleaned['Day_of_Week'] = data_cleaned['Date'].dt.day_name()

# Group by Month and calculate the average sales for each month
avg_sales_by_month = data_cleaned.groupby('Month')['Sales'].mean()

# Group by Day of the Week and calculate the average sales for each day
avg_sales_by_day = data_cleaned.groupby('Day_of_Week')['Sales'].mean()

# Reorder days of the week for logical sequence
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_sales_by_day = avg_sales_by_day.reindex(ordered_days)

# Plotting the results
plt.figure(figsize=(14, 7))

# Plot average sales by month
plt.subplot(1, 2, 1)
avg_sales_by_month.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xticks(ticks=range(len(avg_sales_by_month.index)), 
           labels=pd.to_datetime(avg_sales_by_month.index, format='%m').month_name(), rotation=45)
plt.title('Average Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Plot average sales by day of the week
plt.subplot(1, 2, 2)
avg_sales_by_day.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Average Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
