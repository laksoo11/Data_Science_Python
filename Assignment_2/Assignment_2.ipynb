```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales = pd.read_csv("sales_data.csv")
sales
```

```python
print(sales.isnull().sum())
sales = sales.fillna(sales.mean(numeric_only=True))
print(sales.isnull().sum())
```

```python
sales['Category'] = sales['Category'].fillna(sales['Category'].mode()[0])
print(sales.isnull().sum())
```

```python
data_cleaned=sales.drop_duplicates()
data_cleaned
```

```python
correlation_value = data_cleaned["Price"].corr(data_cleaned["Sales"])
print(f"Correlation between Price and Sales: {correlation_value}")
if correlation_value > 0:
    print("Higher prices tend to correlate with increased sales.")
elif correlation_value < 0:
    print("Higher prices tend to correlate with decreased sales")
else:
    print("There is no correlation between Price and sales.")
```

```python
plt.figure(figsize=(10, 6))

# Adding titles and labels
plt.title("Scatterplot of Price vs Sales", fontsize=14)
plt.xlabel("Price", fontsize=10)
plt.ylabel("Sales", fontsize=10)
#plt.grid(True, linestyle='--', alpha=0.5) 

plt.scatter(data_cleaned["Price"], data_cleaned["Sales"], alpha= 0.4, color = 'r' )
plt.show()


```

```python
# Ensure the 'Date' column is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Sort the data by Date to ensure proper time order
data_cleaned = data_cleaned.sort_values(by='Date')

# Group the data by 'Date' and calculate the sum of sales per day
sales_by_date = data_cleaned.groupby('Date')['Sales'].sum()
sales_by_date
```

```python
# Create a line plot of Sales over time
plt.figure(figsize=(10, 5))
plt.plot(sales_by_date.index, sales_by_date.values, color='g', marker='*')

# Add titles and labels
plt.title("Sales Trend Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
